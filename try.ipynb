```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
excel_data = pd.read_excel('update-data-set.xlsx')
```

```python
data = pd.DataFrame(excel_data, columns=[
                    'date_received', 'product', 'sub_product', 'issue', 'sub_issue', 'consumer_complaint_narr', 'company_public_response', 'company', 'state', 'tags', 'consumer_consent_provided', 'submitted_via', 'date_sent_to_company', 'company_response_to_consumer', 'timely_response', 'consumer_disputed', 'complaint_id'])
```

```python
data.head(50)
```

```python
le = LabelEncoder()
```

```python
data.date_received = le.fit_transform(data.date_received)
#data.product = le.fit_transform(data.product)
data.sub_product = le.fit_transform(data.sub_product)
data.issue = le.fit_transform(data.issue)
data.sub_issue = le.fit_transform(data.sub_issue)
data.consumer_complaint_narr = le.fit_transform(data.consumer_complaint_narr)
data.company_public_response = le.fit_transform(data.company_public_response)
data.company = le.fit_transform(data.company)
data.state = le.fit_transform(data.state)
data.tags = le.fit_transform(data.tags)
data.consumer_consent_provided = le.fit_transform(data.consumer_consent_provided)
data.submitted_via = le.fit_transform(data.submitted_via)
data.date_sent_to_company = le.fit_transform(data.date_sent_to_company)
data.company_response_to_consumer = le.fit_transform(data.company_response_to_consumer)
data.timely_response = le.fit_transform(data.timely_response)
data.consumer_disputed = le.fit_transform(data.consumer_disputed)
data.complaint_id = le.fit_transform(data.complaint_id)
```
